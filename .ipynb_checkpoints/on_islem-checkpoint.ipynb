```python
cv2.line(img,(0,0),(200,512),(0,255,0),3)#başlama durma renk
plt.figure()
plt.imshow(img)
```

```python
cv2.rectangle(img,(0,0),(256,256),(255,0,0),cv2.FILLED)
plt.figure()
plt.imshow(img)
```

```python
cv2.circle(img,(300,300),100,(0,0,255),cv2.FILLED)
plt.figure()
plt.imshow(img)
```

```python
cv2.putText(img,"DİLARA",(325,300),cv2.FONT_HERSHEY_COMPLEX,1,(255,0,0))
plt.figure()
plt.imshow(img)
```

```python
#threshold ekleme kenar tespitlerinde kullanılır belirli yerler öne cıkması için yada
img=cv2.imread("galata.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

(tresh,bfr)=cv2.threshold(img,20,255,cv2.THRESH_BINARY) #20 PİKSEL ÜSTÜNE 255 ATANDI
plt.figure()
plt.imshow(bfr,cmap="gray")
```

```python
#threshold ekleme kenar tespitlerinde kullanılır belirli yerler öne cıkması için yada
img=cv2.imread("galata.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

(tresh,bfr)=cv2.threshold(img,200,255,cv2.THRESH_BINARY) #20 PİKSEL ÜSTÜNE 255 ATANDI
plt.figure()
plt.imshow(bfr,cmap="gray")
```

```python
bfr=cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,5,8) #ayrıntı için
plt.figure()
plt.imshow(bfr,cmap="gray")
```

```python
# blur ekleme gizlilik
img=cv2.imread("galata.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.figure()
plt.imshow(img,cmap="gray")
```
